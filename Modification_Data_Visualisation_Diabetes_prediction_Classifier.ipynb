# **5. Distribution of the Target Variables :**

Visualizing the distribution of target variables is crucial for understanding their importance in data analysis. It provides insights into patterns, variations, and informs decision-making, enhancing overall data comprehension.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='class', data=df)
plt.title('Distribution of Classes')
plt.show()
```

# **6. Boxplot for knowing values size**




This Python method makes use of Seaborn to generate boxplots for specific features in a DataFrame (df). The first boxplot depicts the distribution of the variables 'Age' and 'Pregnancies.' The following plots show the distributions of 'PG_Concentration' with 'Diastolic_BP' and 'Serum_Ins,' 'BMI,' and 'DP_Function.' These visualizations reveal the data's central trend and variability.

```python
sns.boxplot(data=df[['Age', 'Pregnancies']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['PG_Concentration','Diastolic_BP']])
plt.title('Box Plot of Age and feature2')
plt.show()
sns.boxplot(data=df[['Serum_Ins','BMI','DP_Function']])
plt.title('Box Plot of Age and feature2')
plt.show()
```

# **7. Bar Graph**



A bar graph uses rectangular bars of variable lengths to visually display categorical data. Each bar represents a different category, and its length denotes the quantity or frequency. This sort of graph is useful for comparing values across categories, allowing for easy interpretation and analysis of data trends.

```python
X = df.drop('class', axis=1)
y = df['class']

rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X, y)

feature_importances = pd.Series(rf_classifier.feature_importances_, index=X.columns)
feature_importances.sort_values(ascending=False, inplace=True)

plt.figure(figsize=(10, 8))
feature_importances.plot(kind='bar')
plt.title('Feature Importances')
plt.show()
```

# **8. Scatter Plots**


A scatter plot depicts the association between the variables 'Age' and 'Diastolic_BP' in the DataFrame 'df.'

 This code, which makes use of Seaborn, provides a clear representation of individual data points, assisting in the identification of trends or correlations between age and diastolic blood pressure.

  The graphic depicts the influence and relationship of the two variables.

```python
sns.scatterplot(x='Age', y='Diastolic_BP', data= df)
plt.title('Scatter Plot of Age vs Diastolic_BP')
plt.show()
```

# **9. Correlation Matrix**

Using Seaborn, the code builds a correlation matrix for features in the DataFrame 'df'. Correlations are visualized using a heatmap, with warmer colors suggesting stronger associations.

This good representation facilitates in the identification of feature dependencies, which is essential for understanding data linkages and driving analytical findings.

```python
correlation_matrix = df.corr()
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# **10. Stacked Histogram**
Seaborn is used in this code to generate a stacked histogram. Two histograms, 'Tri_Fold_Thick' and 'Diastolic_BP,' with separate colors, are overlaid.

The alpha parameter governs transparency, whereas the legend improves clarity. This visual representation allows you to compare the distributions of two variables inside the same plot.  


```python
sns.histplot(data=df, x='Tri_Fold_Thick', bins=20, color='yellow', alpha=0.9, label='Tri_Fold_Thick')
sns.histplot(data=df, x='Diastolic_BP', bins=22, color='red', alpha=0.9, label='Diastolic_BP')
plt.title('Histogram of Tri_Fold_Thick and Diastolic_BP')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# **11. Pair Plot**


For the selected columns ('DP_Function,' 'Tri_Fold_Thick,' 'Diastolic_BP,' and 'PG_Concentration') the code generates a pair plot using Seaborn.

 This visual representation includes scatter plots for variable pairs and histograms for individual variables, which provide insights into distributions and correlations and aid in thorough exploratory data analysis.

```python
selected_columns = ['DP_Function',	'Tri_Fold_Thick',	'Diastolic_BP',	'PG_Concentration']
selected_df = df[selected_columns]
sns.pairplot(selected_df)
plt.show()
```

#  **12. Combined Violin and Swarm plot**

Using Seaborn and Matplotlib, this code combines a violin plot and a swarm plot.

The violin plot depicts the distribution of 'Age' across 'Tri_Fold_Thick' categories (0: Non-diabetic, 1: Diabetic), whereas the swarm plot overlays individual data points to provide a more thorough picture of age distribution in diabetic and non-diabetic groups.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Tri_Fold_Thick', y='Age', data=df, inner=None, color='lightgray')
sns.swarmplot(x='Tri_Fold_Thick', y='Age', data=df, color='purple', alpha=0.7)
plt.title('Age Distribution by Diabetes Tri_Fold_Thick')
plt.xlabel('Tri_Fold_Thick (0: Non-diabetic, 1: Diabetic)')
plt.ylabel('Age Level')
plt.show()
```

# **13. Scatterplot with Marginal Histograms**


The code creates a "scatterplot with marginal histograms" by employing Seaborn's jointplot. It shows the correlation between 'Diastolic_BP' and 'BMI,' with individual points colored by 'Serum_Ins.' The map reveals the connection between these variables as well as their separate distributions, improving understanding of their interrelationships.

```python
# Assuming 'Serum_Ins' is the target variable
sns.jointplot(x='Diastolic_BP', y='BMI', data=df, hue='Serum_Ins', kind='scatter', palette='coolwarm')
plt.suptitle('Joint Plot of Diastolic_BP and BMI by Diabetes Serum_Ins')
plt.show()
```

# **14. Pairplot to represent specific relations**


The code employs Seaborn to generate a pair plot with points colored by 'Age' for specified features ('DP_Function,' 'Tri_Fold_Thick,' 'Diastolic_BP,' and 'Age.') This graphic sheds light on the correlations and distributions of these attributes, allowing for more in-depth exploratory data analysis.

```python
selected_features = ['DP_Function','Tri_Fold_Thick',	'Diastolic_BP','Age']

custom_palette = sns.color_palette("husl", n_colors=len(df['Age'].unique()))

sns.pairplot(df[selected_features], hue='Age', markers=['D', 'X'],palette = custom_palette)
plt.suptitle('Pairplot of Selected Features', y=1.02, ha='center', fontsize='large')
plt.show()
```
